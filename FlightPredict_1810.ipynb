```python
#테스트 세트 생성
def split_flight_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    flight_indices = shuffled_indices[test_set_size:]
    return data.iloc[flight_indices], data.iloc[test_indices]
```

```python
# 무작위로 샘플 선택하여 데이터셋의 20% 분리
flight_set, test_set = split_flight_test(data_1810, 0.2)
print(len(flight_set), "flight +", len(test_set), "test")
```

```python
from sklearn.model_selection import train_test_split
flight_set, test_set = train_test_split(data_1810, test_size=0.2, random_state=42)
```

```python
#계층별로 데이터셋에 충분한 샘플 수가 있어야함 -> 중요도 편향 방지
data_1810["delay"] = data_1810['DelayRate']
data_1810["delay"].where(data_1810["DelayRate"] < 5, 5.0, inplace=True)
```

```python
#계층별 샘플링
from sklearn.model_selection import StratifiedShuffleSplit
 
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for flight_index, test_index in split.split(data_1810, data_1810["delay"]):
    strat_flight_set = data_1810.loc[flight_index]
    strat_test_set = data_1810.loc[test_index]
```

```python
#전체 데이터셋에서 delay카테고리의 비율측정
data_1810["delay"].value_counts() / len(data_1810)
```

```python
#delay특성 삭제 데이터 원래 상태로 복구
for set_ in (strat_flight_set, strat_test_set):
    set_.drop("delay", axis=1, inplace=True)
```

```python
#회귀 모델의 RMSE
from sklearn.metrics import mean_squared_error
flight_delay_predictions = lin_reg.predict(data_1810)
lin_mse = mean_squared_error(data_1810_labels, flight_delay_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
#DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor
 
tree_reg = DecisionTreeRegressor()
tree_reg.fit(data_1810, data_1810_labels)
```

```python
data_1810_predictions = tree_reg.predict(data_1810)
tree_mse = mean_squared_error(data_1810_labels, data_1810_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
#k-fold cross-validation
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, data_1810, data_1810_labels, scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
#결정트리 점수
display_scores(tree_rmse_scores)
```
